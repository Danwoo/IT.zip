```python
def extract_words_before_postposition(sentence):
    a = kiwi.tokenize(sentence)  # 문장 형태소 분석
    result = []
    for i in range(len(a)):
        if a[i][1] == 'JX' and (a[i][0] == '은' or a[i][0] == '는') and (a[i-1][1] == 'NNP' or a[i-1][1] == 'NNG') and (a[i-1][0] != '사실' and a[i-1][0] != '것'):  # 보조사(JX)와 명사(N으로 시작) 조합
            key = a[i-1][0]
            if a[i-2]:
                if a[i-2][1] == 'NNP' or a[i-2][1] == 'NNG':
                    key = a[i-2][0] + ' ' + key               
            result.append(key)

    return result
```

```python
import copy

summary = ["IBM은 433퀀텀비트의 프로세서를 갖춘 최대 규모의 양자컴퓨터를 개발하는 기록을 남겼고 그보다 훨씬 더 큰 목표를 세웠으며 10년 안에 10만 큐비트의 양자컴퓨터를 개발하는 것이다.",
           "양자컴퓨터는 기존 컴퓨터보다 훨씬 빠른 계산 속도와 더 많은 데이터 처리 능력을 제공한다.","IBM은 상보성 금속 산화물 반도체 기술을 기반으로 하는 집적회로를 저온의 큐비트 옆에 설치하여 수십 밀리와트만 가지고도 큐비트를 제어할 수 있음을 보여주는 원리증명 실험을 수행한 바 있다."]
results = []
summary_list = []
keyword_list = []
masked_list = []

for s in summary:
    keywords = kw_model.extract_keywords([s], keyphrase_ngram_range=(1, 2), stop_words=None, top_n=1)
    keywords = keywords[0][0]
    keyword_masked = copy.deepcopy(keywords)
    tokens = kiwi.tokenize(keywords)

    # 키워드 중에 명사가 아니면 모두 ,로 대체함
    for token in tokens:
        if not token.tag.startswith('N'):                  # 품사 선택!
            keywords = keywords.replace(token.form, ",").strip()
    # , 기준으로 나눈다.
    keywords = keywords.split(",")

    # 공백을 제거하고
    for k in keywords:
        keyword_list.append(k.strip())
    # 리스트 안의 ''를 제거한다.
    for ws in range(keyword_list.count('')):
        keyword_list.remove('')

    # masking sentence
    masked_sentence = s
    for idx, key in enumerate(keyword_list, 1):
        masked_sentence = masked_sentence.replace(key, f'({idx})')
    summary_list.append(s)
    masked_list.append(masked_sentence)

results.append(summary_list)
results.append(keyword_list)
results.append(masked_list)
results
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

def extract_keywords(sentence):
    # 문장을 리스트 형태로 변환 (TfidfVectorizer의 입력 형식)
    sentences = [sentence]
    
    # TfidfVectorizer 객체 생성
    vectorizer = TfidfVectorizer()
    
    # 문장 리스트에 대한 TF-IDF 계산
    tfidf_matrix = vectorizer.fit_transform(sentences)
    
    # TF-IDF 값들을 1차원 배열로 변환
    tfidf_values = np.array(tfidf_matrix.todense()).ravel()
    
    # 중간값 계산
    median_tfidf = np.median(tfidf_values)
    
    # 중간값에 해당하는 단어 추출
    keywords = [word for word, tfidf in zip(vectorizer.get_feature_names_out(), tfidf_values) if tfidf == median_tfidf]
    
    return keywords


sentence = '지구 북반구 지역에서 밤에도 빛을 발해 하늘을 밝히는 야광운 현상이 점점 강해지고 있는데 대기 오염의 결과로 일각에서는 지구 멸망의 전조 현상이 아니냐는 불안감이 고조되고 있다.'
keywords = extract_keywords(sentence)
print(keywords)
```
