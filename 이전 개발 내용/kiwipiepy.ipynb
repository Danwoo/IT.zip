### 사용자 단어 추가

```python
from kiwipiepy import Kiwi

# 사용자 단어 추가 없이 분석해보겠습니다.

kiwi = Kiwi()

print(*kiwi.analyze('사람을 골리다', top_n=5), sep='\n')

print(*kiwi.analyze('골리는 사람이다', top_n=5), sep='\n')


# 사용자 단어 '골리'를 추가해보도록 하겠습니다.
kiwi = Kiwi()
kiwi.add_user_word('골리', 'NNP', 0)

print(*kiwi.analyze('사람을 골리다', top_n=5), sep='\n')

print(*kiwi.analyze('골리는 사람이다', top_n=5), sep='\n')

# 사용자 단어 '골리'의 점수를 낮춰서 추가해보도록 하겠습니다.
kiwi = Kiwi()
kiwi.add_user_word('골리', 'NNP', -6)

print(*kiwi.analyze('사람을 골리다', top_n=5), sep='\n')

print(*kiwi.analyze('골리는 사람이다', top_n=5), sep='\n')    
```

```python
kiwi = Kiwi()
a = kiwi.tokenize('정수형 i를 1로 초기화한 뒤 i가 5보다 작거나 같은지 체크, 작거나 같다면 i를 출력하고 i값 1증가 5보다 크다면 종료')
completed_sentences = ['<시작>']
completed_token = []
for token in a:
    if token.tag != 'Z_CODA':
        if token.form == '뒤' or token.form == ',' or token.form == '고' or token.form == '면' or token.form == '종료':
          completed_sentences.append('<SEP>')
          completed_token.append('<SEP>')
          continue
        completed_token.append(token)
        completed_sentences.append(token.form + ' ')
completed_sentences.append(' <종료>')
completed_text = ''.join(completed_sentences)
print(completed_text)
```

```python
parts = []
start = 0
for i in range(len(completed_token)):
  if completed_token[i] == '<SEP>':
    parts.append(completed_token[start:i])
    start += i+1
```

```python
def split_sentences(text):
  kiwi = Kiwi()
  a = kiwi.tokenize(text)
  completed_sentences = ['<시작>']
  completed_token = []
  for token in a:
      if token.tag != 'Z_CODA':
          if token.form == '뒤' or token.form == ',' or token.form == '고' or token.form == '면' or token.form == '종료':
            completed_sentences.append('<SEP>')
            completed_token.append('<SEP>')
            continue
          completed_token.append(token)
          completed_sentences.append(token.form + ' ')
  completed_sentences.append(' <종료>')
  completed_text = ''.join(completed_sentences)
  print(completed_text)
  return completed_text
```
